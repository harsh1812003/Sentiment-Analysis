```python
import pandas as pd
```

```python
df=pd.read_csv('Combined_News_DJIA.csv')
df.head()
```

```python
data=df.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

```python
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from gensim.models import KeyedVectors
```

```python
import gensim.downloader as api

wv = api.load('word2vec-google-news-300')
```

```python
voc_size = 20000  # Vocabulary size
embedding_vector_features = 300  # Word2Vec embedding size
sent_length = 1000  # Maximum length of the sequences
```

```python
tokenizer = Tokenizer(num_words=voc_size)
```

```python
tokenizer.fit_on_texts(headlines)
```

```python
sequences = tokenizer.texts_to_sequences(headlines)
```

```python
X = pad_sequences(sequences, maxlen=sent_length)
```

```python
word_index = tokenizer.word_index
```

```python
embedding_matrix = np.zeros((voc_size, embedding_vector_features))
```

```python
for word, i in word_index.items():
    if i < voc_size:
        try:
            embedding_vector = wv[word]
            embedding_matrix[i] = embedding_vector
        except KeyError:
            pass  
```

```python
model = Sequential()
embedding_layer = Embedding(input_dim=voc_size,
                            output_dim=embedding_vector_features,
                            weights=[embedding_matrix],
                            input_length=sent_length,
                            trainable=False)  # Set trainable=False to not update the embeddings

model.add(embedding_layer)
model.add(LSTM(500))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

print(model.summary())
```

```python
len(headlines)
```

```python
X_train=X[0:1500]
```

```python
y_train=np.array(df['Label'][0:1500])
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=40)
```

```python
X_test=X[1500:]
y_test=np.array(df['Label'][1500:])
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.5, 1,0)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
